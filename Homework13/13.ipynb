```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

# Задание 12


```python
data = pd.DataFrame({"x":[10,  10,  10,  13,  13,  13,  16,  16,  16,  19,  19,  19,  22,  22,  22,  25,  25,  25],
                     "y":[110, 130, 150, 130, 150, 170, 150, 170, 190, 150, 170, 190, 190, 210, 230, 210, 230, 250],
                     "n":[1,     3,   4,   5,   6,   5,   4,   8,   6,   6,  15,   9,   5,   6,   7,   1,   7,   2]})
```

```python
# Расчет среднего для x и y
x_avg = (data["x"] * data["n"]).sum() / data['n'].sum()
y_avg = (data["y"] * data["n"]).sum() / data['n'].sum()
# Расчет сдантартного отклонения для x и y
sigma_x = np.sqrt((data['x'] ** 2 * data['n']).sum() / data['n'].sum() - x_avg ** 2)
sigma_y = np.sqrt((data['y'] ** 2 * data['n']).sum() / data['n'].sum() - y_avg ** 2)
# Расчет среднего для x * y
xy_avg = (data['x'] * data['y'] * data['n']).sum() / data['n'].sum()
# Расчет коэффициента корреляции
r = (xy_avg - x_avg * y_avg) / (sigma_y * sigma_x)
# Расчет коэффициентов регрессии 
a = r * sigma_y / sigma_x
b = y_avg - x_avg * a

x_arr = []
y_arr = []
for ind in data['x'].unique():
    curr = data.loc[data['x'] == ind]
    x_arr.append(ind)
    y_arr.append((curr['y'] * curr['n']).sum() / curr['n'].sum())
plt.scatter(x=data['x'], y=data['y'], c=data['n'],  cmap='cividis')
plt.plot(x_arr, y_arr)
plt.colorbar(label='Частота')
plt.show()
print(f'Уравнение регрессии: y = {a} * x + {b}')
```
